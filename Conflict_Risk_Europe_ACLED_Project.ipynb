````python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

# Load your merged dataset
df = pd.read_csv('merged_file.csv')

# Filter for 20212023
df_filtered = df[df['year'].isin([2021, 2022, 2023])].copy()
df_filtered = df_filtered.dropna()
df_filtered['WEEK'] = pd.to_datetime(df_filtered['WEEK'])
df_filtered['POPULATION_EXPOSURE'] = pd.to_numeric(df_filtered['POPULATION_EXPOSURE'], errors='coerce')

# Categorize protests
df_filtered['IS_VIOLENT'] = df_filtered['EVENT_TYPE'].isin(['Riots', 'Violence against civilians']) | (df_filtered['FATALITIES'] > 0)
df_filtered['DEMO_TYPE'] = np.where(df_filtered['IS_VIOLENT'], 'Violent', 'Peaceful')

# Democracy score
df_filtered['DEMOCRACY_SCORE'] = df_filtered[['representation_est', 'free_express_est', 'civil_lib_est', 'rule_law_est']].mean(axis=1)

# Aggregate results
results = []
for (country, year), group in df_filtered.groupby(['country_name', 'year']):
    weekly_data = group.groupby(pd.Grouper(key='WEEK', freq='W'))
    weekly_peaceful = weekly_data['DEMO_TYPE'].apply(lambda x: (x == 'Peaceful').sum())
    weekly_violent = weekly_data['DEMO_TYPE'].apply(lambda x: (x == 'Violent').sum())
    weekly_total = weekly_peaceful.add(weekly_violent, fill_value=0)
    weekly_violent_freq_percent = (weekly_violent / weekly_total * 100).fillna(0)

    if len(weekly_data) > 1 and weekly_total.nunique() > 1 and weekly_violent_freq_percent.nunique() > 1:
        corr, _ = pearsonr(weekly_violent_freq_percent, weekly_total)
    else:
        corr = np.nan

    peaceful_count = (group['DEMO_TYPE'] == 'Peaceful').sum()
    violent_count = (group['DEMO_TYPE'] == 'Violent').sum()
    total_fatalities = group['FATALITIES'].sum()
    avg_democracy_score = group['DEMOCRACY_SCORE'].mean()
    violent_freq_percent_annual = (violent_count / (peaceful_count + violent_count) * 100) if (peaceful_count + violent_count) > 0 else 0
    total_exposure = group['POPULATION_EXPOSURE'].sum()

    results.append({
        'country_name': country,
        'year': year,
        'PEACEFUL_PROTESTS': peaceful_count,
        'VIOLENT_PROTESTS': violent_count,
        'TOTAL_FATALITIES': total_fatalities,
        'DEMOCRACY_SCORE': avg_democracy_score,
        'VIOLENT_FREQ_PERCENT': violent_freq_percent_annual,
        'CORRELATION': corr,
        'POPULATION_EXPOSURE': total_exposure
    })

# Final table
results_df = pd.DataFrame(results)
print(results_df.to_string(index=False))
````
